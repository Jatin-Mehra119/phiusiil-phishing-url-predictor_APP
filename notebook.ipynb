```python
import pandas as pd 
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
df = pd.read_csv("PhiUSIIL_Phishing_URL_Dataset.csv")
```

```python
data = df[["URL", "TLD", "URLLength", "Domain", "DomainLength", "IsHTTPS", "HasTitle", "Title","NoOfImage", "NoOfCSS", "NoOfJS", "NoOfExternalRef", "label"]]
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(data, test_size=0.2, random_state=42)
train_X, train_y = train.drop("label", axis=1).copy(), train["label"].copy()
test_X, test_y = test.drop("label", axis=1).copy(), test["label"].copy()
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.compose import ColumnTransformer, make_column_selector


Num_pipeline = make_pipeline(StandardScaler())
TDL_pipeline = make_pipeline(OneHotEncoder(handle_unknown='ignore'))

# Preprocessor with ColumnTransformer
preprocessor = ColumnTransformer([
    ('num', Num_pipeline, make_column_selector(dtype_include=np.number)),
    ('tdl', TDL_pipeline, ['TLD'])
], 
remainder='drop') 

preprocessor.fit_transform(train_X)
```

```python
# Full pipeline with preprocessing and model
full_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', RandomForestClassifier())
])

# Fit the pipeline to the training data
full_pipeline.fit(train_X, train_y)
```

```python
from sklearn.metrics import classification_report

# Predictions
Pred_y = full_pipeline.predict(test_X)
print(classification_report(test_y, Pred_y))
```
